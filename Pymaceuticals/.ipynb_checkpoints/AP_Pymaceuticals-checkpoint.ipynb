```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.stats import linregress


# Study data files, path & rename
mouse_metadata_path = Path('data/Mouse_metadata.csv')
study_results_path = Path('data/Study_results.csv')

mouse_metadata_df = pd.read_csv(mouse_metadata_path)
mouse_metadata_df = mouse_metadata_df.rename(columns = {'Age_months' : 'Age (months)'})

study_results_df = pd.read_csv(study_results_path)

# Make a single DataFrame
mice_study_results_df = pd.merge(study_results_df, mouse_metadata_df, how = 'left', on = ['Mouse ID', 'Mouse ID'])
                         
# Combined data frame preview
mice_study_results_df
```

```python
# Number of mice
total_mice_count = len(mice_study_results_df['Mouse ID'].unique())
total_mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
find_duplicate = mice_study_results_df[mice_study_results_df.duplicated(['Mouse ID', 'Timepoint'])].groupby('Mouse ID')
find_duplicate.head()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data = mice_study_results_df.loc[mice_study_results_df['Mouse ID'] == 'g989', :]
duplicate_data
```

```python
# New DataFrame without duplicate data removed
mice_study_results_df = mice_study_results_df.loc[mice_study_results_df['Mouse ID'] != 'g989', :]
clean_study_data_df = mice_study_results_df.reset_index(drop = True)
mice_results_df = clean_study_data_df.copy()
mice_results_df.head(110)
# Verifying that the index has been reset since data removed from index 107
```

```python
# Number of unique mice data
len(mice_results_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.


# Index = Drug Regimen 
# Mean Tumor Volume
# Median Tumor Volume
# Tumor Volume Variance
# Tumor Volume Std. Dev.
# Tumor Volume Std. Err

mean_tum_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean(), 4)
median_tum_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median(), 4)
variance_tum_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var(), 4)
std_dev_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std(), 4)
std_err_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem(), 4)

# print(f'{mean_tum_vol} n\ n\ {median_tum_vol} n\n\ {variance_tum_vol} n\n\ {std_dev_vol} n\n\ {std_err_vol}')

summary_tumor_df = pd.DataFrame({
    'Tumor Vol. Mean': mean_tum_vol,
    'Tumor Vol. Median': median_tum_vol,
    'Tumor Vol. Variance': variance_tum_vol,
    'Tumor Vol. Std. Dev': std_dev_vol,
    'Tumor Vol. Std. Err': std_err_vol})
summary_tumor_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# whoops I did both!

# Using the aggregation method, produce the same summary statistics in a single line of code
tumor_summary_df = round(
    mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg([
        'mean', 'median', 'var', 'std', 'sem'], axis = 1), 
    4)
tumor_summary_df
```

```python
final_tumor_vol_df['Tumor Volume (mm3)'] = final_tumor_vol_df['Tumor Volume (mm3)'].astype(int)
final_tumor_vol_df.dtypes
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimen = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

for drug in regimen:
    final_tumor_vol_by_drug = final_tumor_vol_df.loc[final_tumor_vol_df['Drug Regimen'] == drug, 'Tumor Volume (mm3)']
    tumor_vol.append(final_tumor_vol_by_drug)

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = final_tumor_vol_by_drug.quantile([.25,.5,.75])
    lowq = quartiles[0.25]
    upq = quartiles[0.75]
    iqr = upq-lowq
    lower_bound = lowq - (1.5*iqr)
    upper_bound = upq + (1.5*iqr)
    
    outliers = final_tumor_vol_by_drug.loc[(final_tumor_vol_by_drug < lower_bound) | (final_tumor_vol_by_drug > upper_bound)]
    
    #won't work why?
    
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.boxplot(tumor_vol, labels = regimen, flierprops = dict( marker = 'o', markersize =  10, markerfacecolor = 'aqua'))
ax1.set_ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

avg_tumor_cap_vol = capomulin_mice_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weight = capomulin_mice_df.groupby('Mouse ID')['Weight (g)'].mean()


plt.scatter(mouse_weight,avg_tumor_cap_vol, color = 'plum')

plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python


# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen


# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

correlation = st.pearsonr(mouse_weight,avg_tumor_cap_vol)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],3)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(mouse_weight, avg_tumor_cap_vol)
regress_values = mouse_weight * slope + intercept
plt.scatter(mouse_weight,avg_tumor_cap_vol, color = 'teal')
plt.plot(mouse_weight,regress_values,"r-", color = 'lightgreen')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()

```
