```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st


# Study data files, path & rename
mouse_metadata_path = Path('data/Mouse_metadata.csv')
study_results_path = Path('data/Study_results.csv')

mouse_metadata_df = pd.read_csv(mouse_metadata_path)
mouse_metadata_df = mouse_metadata_df.rename(columns = {
    'Age_months' : 'Age (months)'})

study_results_df = pd.read_csv(study_results_path)

# Make a single DataFrame
mice_study_results_df = pd.merge(study_results_df, mouse_metadata_df, how = 'left', on = ['Mouse ID', 'Mouse ID'])
                         
# Combined data frame preview
mice_study_results_df
```

```python
# Number of mice
total_mice_count = len(mice_study_results_df['Mouse ID'].unique())
total_mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
find_duplicate = mice_study_results_df[mice_study_results_df.duplicated(['Mouse ID', 'Timepoint'])].groupby('Mouse ID')
find_duplicate.head()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data = mice_study_results_df.loc[mice_study_results_df['Mouse ID'] == 'g989', :]
duplicate_data
```

```python
# New DataFrame without duplicate data removed
mice_study_results_df = mice_study_results_df.loc[mice_study_results_df['Mouse ID'] != 'g989', :]
clean_study_data_df = mice_study_results_df.reset_index(drop = True)
mice_results_df = clean_study_data_df.copy()
mice_results_df.head(110)
# Verifying that the index has been reset since data removed from index 107
```

```python
# Number of unique mice data
len(mice_results_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.


# Index = Drug Regimen 
# Mean Tumor Volume
# Median Tumor Volume
# Tumor Volume Variance
# Tumor Volume Std. Dev.
# Tumor Volume Std. Err

mean_tum_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean(), 4)
median_tum_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median(), 4)
variance_tum_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var(), 4)
std_dev_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std(), 4)
std_err_vol = round(mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem(), 4)

# print(f'{mean_tum_vol} n\ n\ {median_tum_vol} n\n\ {variance_tum_vol} n\n\ {std_dev_vol} n\n\ {std_err_vol}')

summary_tumor_df = pd.DataFrame({
    'Tumor Vol. Mean': mean_tum_vol,
    'Tumor Vol. Median': median_tum_vol,
    'Tumor Vol. Variance': variance_tum_vol,
    'Tumor Vol. Std. Dev': std_dev_vol,
    'Tumor Vol. Std. Err': std_err_vol})
summary_tumor_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# whoops I did both!

# Using the aggregation method, produce the same summary statistics in a single line of code
tumor_summary_df = round(
    mice_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg([
        'mean', 'median', 'var', 'std', 'sem'], axis = 1), 
    4)
tumor_summary_df
```

## Bar and Pie Charts

```python
# Bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas
mice_timepoints_bar_pandas = mice_results_df.groupby('Drug Regimen')['Timepoint'].count().plot.bar(
    ylabel ='# of Observed Mouse Timepoints', color = 'plum')
mice_timepoints_bar_pandas
```

```python
# Bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
## mice_time_count = mice_results_df.groupby('Drug Regimen')['Timepoint'].count().to_xarray()
timepoint_count = mice_results_df.groupby('Drug Regimen')['Timepoint'].count().tolist()
x_axis = len(timepoint_count)
# timepoint_count
drug_list = summary_tumor_df.index.tolist()
# drug_list
bin_width = 0.5

plt.bar(drug_list, timepoint_count, width = bin_width, color = 'plum', alpha = 1, align = 'center')
plt.xlim(-.55, x_axis-0.4)
plt.xlabel('Drug Regimen')
plt.xticks(rotation = 90)
plt.ylabel('# of Observed Mouse Timepoints')
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
mice_gender_pie_pandas = mice_results_df.groupby('Sex')['Sex'].count().plot.pie(
    autopct='%1.1f%%',
    explode=[0.005, 0.005],
    colors = ['orchid','darkturquoise'],
    startangle = 180)
mice_gender_pie_pandas
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
# gender_count = mice_results_df[mice_results_df.groupby('Sex')].count()
sizes_pyplot = mice_results_df.groupby('Sex')['Sex'].count().to_xarray()
labels_pyplot = mice_results_df.groupby('Sex')['Sex'].count().index.tolist()
colors = ['orchid','darkturquoise']

plt.ylabel('Sex')
plt.pie(
    sizes_pyplot,
    explode = (0.005, 0.005),
    labels = labels_pyplot,
    colors = colors,
    autopct='%1.1f%%', 
    shadow = False,
    startangle = 180)
plt.axis('equal')
plt.show()
```
